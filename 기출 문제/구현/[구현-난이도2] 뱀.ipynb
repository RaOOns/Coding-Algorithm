# **뱀**

사과를 먹으면 뱀 길이가 늘어나는 게임이 있다.

게임은 벽 또는 자기 자신의 몸과 부딪히면 게임이 끝난다.

게임음 N x N 정사각 보드 위에서 진행되고, 몇몇 칸에는 사과가 놓여져 있다.

보드의 상하좌우 끝에는 벽이 있으며, 게임 시작 시 뱀은 맨 위 맨 좌측에 위치하고 뱀의 길이는 1이다.

뱀은 처음에 오른쪽으로 이동한다.

- 뱀은 몸 길이를 늘려 머리를 다음 칸에 위치
- 만약 이동한 칸에 사과가 있다면, 그 칸에 있던 사과가 없어지고 꼬리는 움직이지 않음
- 만약 이동한 칸에 사과가 없다면, 몸길이를 줄여서 꼬리가 위치한 칸을 비워준다. 즉, 몸 길이는 변하지 않는다.

사과의 위치와 뱀의 이동 경로가 주어질 때 이 게임이 몇 초에 끝나는지 계산하시오.

```python
n = int(input())

# 게임 맵
space = [[0] * n for _ in range(n)]
space[0][0] = 1

# 사과 위치를 2로 표시
for _ in range(int(input())):
    apple_x, apple_y = map(int, input().split())
    space[apple_x-1][apple_y-1] = 2

# 이동 경로 정보 저장
route = {}
for _ in range(int(input())):
    time, direct = input().split()
    route[int(time)] = direct


# 상우하좌
dx = [-1, 0, 1 ,0]
dy = [0, 1, 0, -1]


time = 0            # 게임 시간
now_x, now_y = 0, 0 # 현재 머리 위치
snake = [[0,0]]     # 뱀이 차지하는 공간
direct = 1          # 이동 방향

while True:

    # 1초 후 direct 방향으로 1칸 이동
    time += 1
    now_x += dx[direct]
    now_y += dy[direct]

    # 다음 조건이면 게임 종료
    if now_x < 0 or now_x >= n or now_y < 0 or now_y >= n or space[now_x][now_y] == 1:
        print(time)
        break

    # 이동 경로에 사과가 있으면, 뱀 크기 증가
    elif space[now_x][now_y] == 2:
        space[now_x][now_y] = 1
        snake.append([now_x, now_y])            # 뱀이 차지하는 공간 확장

    # 이동 경로에 사과가 없으면, 뱀 꼬리 이동
    elif space[now_x][now_y] == 0:
        space[now_x][now_y] = 1
        snake.append([now_x, now_y])            # 뱀이 차지하는 공간 변환
        space[snake[0][0]][snake[0][1]] = 0     # 뱀 꼬리 이동
        snake.pop(0)                            # 뱀 꼬리가 차지하던 공간 제거

    # print(snake)1
    # 회전할 시간인 경우 회전
    if time in route.keys():
        if route[time] == "L":
            direct = (direct - 1) % 4
        elif route[time] == "D":
            direct = (direct + 1) % 4

    # print("=="*10); display(space)
```
