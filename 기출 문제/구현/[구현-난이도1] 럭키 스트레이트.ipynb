# **럭키 스트레이트**

특정 조건에서만 사용 가능한 게임 기술인 럭키 스트레이트가 있다.

특정 조건은 캐릭터의 점수 N에 대하여 자릿수를 기준으로 왼쪽 부분의 각 자릿수의 합과 오른쪽 부분의 각 자릿수의 합이 동일한 상황이다.

예를들어 123,402라면 왼쪽은 1+2+3, 오른쪽은 4+0+2이므로 두 합이 6으로 기술을 사용할 수 있다.

럭키 스트레이트를 사용할 수 있다면 'LUCKY' 아니라면 'READY'를 출력

ex) 123402, 7755

```python
data = input()

mid = int(len(data)/2)

left = 0
right = 0

for le in data[:mid]:
    left += int(le)

for ri in data[mid:]:
    right += int(ri)

if left == right:
    print("LUCKY")
else:
    print("READY")
```

```python
data = input()

mid = int(len(data)/2)

left = 0
right = 0

for i in range(len(data)):
    if i < mid:
        left += int(data[i])
    else:
        right += int(data[i])

if left == right:
    print("LUCKY")
else:
    print("READY")
```
