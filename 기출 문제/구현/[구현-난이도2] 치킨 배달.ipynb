# **치킨 배달**

크기가 N x N인 도시가 있다. 도시는 1x1 크기의 칸으로 나누어져 있다.

도시의 각 칸은 빈칸, 치킨집, 집 중 하나이다.

도시의 칸은 (r, c)와 같은 형태로 나타내고, r행 c열 또는 위에서부터 r번째 칸, 왼쪽부터 c번째 칸을 의미한다. (r과 c는 1부터 시작)

"치킨 거리"는 집과 가장 가까운 치킨집 사이의 거리이며, 집을 기준으로, 각각의 집은 치킨 거리를 가지고 있다.

"도시의 치킨 거리"는 모든 집의 치킨 거리의 합이다.

(r1, c1)과 (r2, c2) 사이의 거리는 |r1 - r2| + |c1 - c2|로 구한다.

0은 빈칸, 1은 집, 2는 치킨집을 의미

도시의 치킨 거리가 가장 작게 되는 최대 M개의 치킨집을 고르고, 이때 도시의 치킨 거리를 구하시오.

```python
n, m = map(int, input().split())

home = []
chicken = []

for i in range(n):
    data = list(map(int, input().split()))
    for j in range(n):
        if data[j] == 1:
            home.append([i+1, j+1])
        elif data[j] == 2:
            chicken.append([i+1, j+1])

from itertools import combinations

select = list(combinations(chicken, m))

distance = []
for s in select:        # 선택된 M개의 치킨집
    dist = 0
    for h in home:      # 각 집에 대해 각 치킨집까지의 거리 계산 후 가장 가까운 거리를 선택해 더함
        d = int(1e9)
        for c in s:
            d = min(d, abs(c[0] - h[0]) + abs(c[1] - h[1]))
        dist += d
    distance.append(dist)    # 선택된 M개의 치킨집에 대한 도시의 치킨 거리

distance.sort()

print(distance[0])
```
