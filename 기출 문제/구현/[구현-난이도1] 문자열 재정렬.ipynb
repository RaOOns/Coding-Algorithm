# **문자열 재정렬**

알파벳 대문자와 숫자(0~9)로만 구성된 문자열이 입력으로 주어진다.

이때 모든 알파벳을 오름차순으로 정렬하여 이어서 출력한 뒤, 모든 숫자를 더한 값을 이어서 출력.

ex) K1KAC5CB7  ->  ABCKK13

```python
data = input()

char = []
num = 0

for i in data:
    if i.isalpha():
        char.append(i)
    else:
        num += int(i)

char.sort()

if num != 0 :
    char.append(str(num))

print("".join(char))
```
