# **기둥과 보 설치**

죠르디는 기둥과 보를 이용하여 벽면 구조물을 자동으로 세우는 로봇을 개발할 계획이다.

프로그램은 2차원 가상 벽면에 기둥과 보를 이용한 구조물을 설치 가능하며, 기둥과 보는 길이가 1인 선분으로 표현되며 다음과 같은 규칙이 있다.
- 기둥은 바닥 위에 있거나 보의 한쪽 끝부분 위에 있거나, 또는 다른 기둥 위에 있어야 한다.
- 보는 한쪽 끝부분이 기둥 위에 있거나, 또는 양쪽 끝부분이 다른 보와 동시에 연결되어 있어야 한다.

2차원 벽면은 n x n 크기 정사각 격자 형태이며, 각 격자는 1x1 크기이다.

맨 처음 벽면은 비어 있다.

기둥과 보는 격자 선의 교차점에 걸치지 않고, 격자 칸의 각 변에 정확히 일치하도록 설치 가능

https://school.programmers.co.kr/learn/courses/30/lessons/60061

```python
# 현재 설치된 구조물이 가능한지 확인
def possible(answer):
    for x, y, stuff in answer:
        if stuff == 0:          # 설치물이 기둥인 경우
            # '바닥 위' or '보의 한쪽 끝부분 위' or '다른 기둥 위'라면 정상
            if  y == 0 or [x-1, y, 1] in answer or [x, y, 1] in answer or [x, y-1, 0] in answer:
                continue

            # 아니라면 거짓
            return False

        elif stuff == 1:            # 설치물이 보인 경우
        # '한쪽 끝부분이 기둥 위' or '양쪽 끝부분이 다른 보와 동시에 연결'이라면 정상
            if [x, y-1, 0] in answer or [x+1, y-1, 0] in answer or ([x-1, y, 1] in answer and [x+1, y, 1] in answer):
                continue

            # 아니라면 거짓
            return False

    return True


def solution(n, build_frame):
    answer = []

    for frame in build_frame:   # 작업(frame)의 개수는 최대 1,000개
        x, y, stuff, operate = frame

        # 삭제하는 경우
        if operate == 0:
            answer.remove([x, y, stuff])        # 일단 삭제
            if not possible(answer):            # 가능한 구조인지 확인
                answer.append([x, y, stuff])    # 아니라면 다시 설치

        # 설치하는 경우
        if operate == 1:
            answer.append([x, y, stuff])        # 일단 설치
            if not possible(answer):            # 가능한 구조인지 확인
                answer.remove([x, y, stuff])    # 아니라면 삭제

    return sorted(answer)
```

```python
print(solution(5, [[1,0,0,1],[1,1,1,1],[2,1,0,1],[2,2,1,1],[5,0,0,1],[5,1,0,1],[4,2,1,1],[3,2,1,1]]))
print("=="*30)
print(solution(5, [[0,0,0,1],[2,0,0,1],[4,0,0,1],[0,1,1,1],[1,1,1,1],[2,1,1,1],[3,1,1,1],[2,0,0,0],[1,1,1,0],[2,2,0,1]]))
```
