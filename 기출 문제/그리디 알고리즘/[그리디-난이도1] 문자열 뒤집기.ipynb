# **문자열 뒤집기**

다솜이는 0과 1로 이루어진 문자열 S를 가지고 있다.

다솜이는 문자열 S에 있는 모든 문자를 전부 같게 만들려고 한다.

할 수 있는 행동은 S에서 연속된 하나 이상의 숫자를 잡고 모두 뒤집는 것.

문자열이 주어졌을 때, 다솜이가 해야 하는 행동의 최소 횟수

ex) 0001100

```python
data = input()

zero = 0
one = 0

count = [int(data[0])]
for i in data[1:]:

    if count[-1] == int(i):
        count.append(int(i))
    else :
        if count[-1] == 0:
            zero += 1
        else:
            one += 1
        count = [int(i)]

result = min(zero, one)

print(result)
```
