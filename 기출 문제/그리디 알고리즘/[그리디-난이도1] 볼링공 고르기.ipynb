# **볼링공 고르기**

A, B 두 사람이 볼링을 치고 있다. 두 사람은 서로 무게가 다른 볼링공을 고르려고 한다.

볼링공은 총 N개가 있으며 각 볼링공마다 무게가 적혀 있고, 공의 번호는 1번부터 순서대로 부여된다.

똑같은 무게의 공이 여러 개 있을 수 있지만, 서로 다른 공으로 간주.

볼링공의 무게는 1부터 M까지의 자연수 형태로 존재

두 사람이 서로 다른 볼링공을 고르는 경우의 수를 구하는 프로그램을 작성



```python
n, m = map(int, input().split())
balls = list(map(int, input().split()))

result = 0

for ball in balls:
    temp = [i for i in balls if i != ball ]
    result += len(temp)

print(int(result/2))
```

```python
n, m = map(int, input().split())
balls = list(map(int, input().split()))


# 1부터 10까지의 무게를 담을 수 있는 리스트
array = [0] * 11; print(array)

# 각 무게에 해당하는 볼링공의 개수 카운트
for x in balls:
    array[x] += 1
print(array)


result = 0
# 1부터 m까지의 각 무게에 대하여 처리
for i in range(1, m+1):
    n -= array[i]               # 무게가 i인 볼링공의 개수(A가 선택할 수 있는 개수) 제외
    result += array[i] * n      # B가 선택하는 경우의 수와 곱하기

print(result)
```
