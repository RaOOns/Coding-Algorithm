# **만들 수 없는 금액**


N개의 동전이 있다. 이때, N개의 동전을 이용하여 만들 수 없는 양의 정수 금액 중 최솟값을 구하는 프로그램을 작성

예를 들어, N = 5이고, 각 동전이 3원, 2원, 1원, 1원, 9원짜리 동전이라면 만들 수 없는 최솟값은 8원이다.

```python
n = int(input())
unit = list(map(int, input().split()))

unit.sort()

target = 1

for x in unit:
    if target < x:
        break
    target += x

print(target)
```
