# **정수 삼각형**

정수 삼각형의 맨 위의 값부터 시작해서 아래에 있는 수 중 하나를 선택해 아래층으로 내려올 때, 이제까지 선택된 수의 합이 최대가 되는 경로를 구하는 프로그램 작성

아래층에 있는 수는 현재 층에서 선택된 수의 대각선 아래 또는 대각선 오른쪽에 있는 것 중에서만 선택 가능

```python
n = int(input())
data = []
for _ in range(n):
    data.append(list(map(int, input().split())))

for i in range(1, n):
    for j in range(i+1):
        # 왼쪽 위에서 내려오는 경우
        if j == 0:
            up_left = 0
        else:
            up_left = data[i-1][j-1]

        # 바로 위에서 내려오는 경우
        if j == i:
            up = 0
        else:
            up = data[i-1][j]

        # 최대 합을 저장
        data[i][j] = data[i][j] + max(up_left, up)

print(max(data[n-1]))
```
