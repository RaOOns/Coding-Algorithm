# **편집 거리**

두 개의 문자열 A와 B가 주어졌을 대, 문자열 A를 편집하여 문자열 B로 만들고자 한다.

문자열 A을 편집할 때는 다음의 세 연산 중에서 한 번에 하나씩 선택하여 이용할 수 있다.

1. 삽입: 특정한 위치에 하나의 문자를 삽입
2. 삭제: 특정한 위치에 있는 하나의 문자를 삭제
3. 교체: 특정한 위치에 있는 하나의 문자를 다른 문자로 교체

편집 거리란 문자열 A를 편집하여 문자열 B로 만들기 위해 사용한 연산의 수를 의미

문자열 A를 문자열 B로 만드는 최소 편집 거리를 계산하는 프로그램을 작성하세요.

예를 들어 sunday와 saturday의 최소 편집 거리는 3이다.

```python
a = input()
b = input()

consist_a = list(a)
consist_b = list(b)

result = 0
# 두 문자열의 길이가 같다면, 교체 사용
if len(a) == len(b):
    result += len(set(consist_a) - set(consist_b))  # 서로 다른 문자 개수만큼 교체

# 두 문자열의 길이가 다르다면, 삽입 또는 제거 후 교체
elif len(a) != len(b):
    com = len(set(consist_a) & set(consist_b))  # 겹치는 갯수
    result += max(len(a), len(b)) - com         # 겹치는 것 빼고 전부 삽입, 삭제, 교체

print(result)
```
